# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

#import numpy as np

# File to Load (Remember to change these)
city_data_to_load = 'Resources/city_data.csv'
ride_data_to_load = 'Resources/ride_data.csv'

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head(10)
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
#city_type = pd.Series(['Rural', 'Suburban', 'Urban'])
#print(city_type)

total_rides_group = pyber_data_df.groupby(['type']).count()["ride_id"]
total_rides_group
```

```python
# 2. Get the total drivers for each city type
total_drivers = city_data_df.groupby(['type']).sum()["driver_count"]

total_drivers
```

```python
#  3. Get the total amount of fares for each city type
total_amount_fares = pyber_data_df.groupby(['type']).sum()["fare"]

total_amount_fares
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_ride = total_amount_fares / total_rides_group

average_fare_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_driver = total_amount_fares / total_drivers

average_fare_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_data_df = pd.DataFrame(
        {"Total Rides": total_rides_group,
        "Total Drivers": total_drivers,
        "Total Fares": total_amount_fares,
        "Average Fare per Ride": average_fare_ride,
        "Average Fare per Driver": average_fare_driver})
pyber_data_df 
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_data_df.index.name = None
```

```python
#  8. Format the columns.
pyber_data_df["Total Rides"] = pyber_data_df["Total Rides"].map("{:.0f}".format)
pyber_data_df["Total Drivers"] = pyber_data_df["Total Drivers"].map("{:.0f}".format)
pyber_data_df["Total Fares"] = pyber_data_df["Total Fares"].map("${:.2f}".format)
pyber_data_df["Average Fare per Ride"] = pyber_data_df["Average Fare per Ride"].map("${:.2f}".format)
pyber_data_df["Average Fare per Driver"] = pyber_data_df["Average Fare per Driver"].map("${:.2f}".format)

pyber_data_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

type_sum_fare = pyber_data_df.groupby(["type", "date"]).sum()[["fare"]]
type_sum_fare.tail(10)
# print(type(type_sum_fare))
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
type_sum_fare = type_sum_fare.reset_index()
type_sum_fare.tail(5)
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
type_sum_fare_pivot = type_sum_fare.pivot(index="date", columns="type", values="fare")
type_sum_fare_pivot.tail(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
given_date_country_sumfare = type_sum_fare_pivot.loc["2019-01-01":"2019-04-29"]
given_date_country_sumfare.head(20)
#given_date_country_sumfare.info()
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
given_date_country_sumfare.index = pd.to_datetime(given_date_country_sumfare.index)
given_date_country_sumfare
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
given_date_country_sumfare.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
week_country_sumfare = given_date_country_sumfare.resample("W").sum()
week_country_sumfare.tail(20)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

Rural_value = week_country_sumfare['Rural']
Suburban_value = week_country_sumfare['Suburban']
Urban_value = week_country_sumfare['Urban']

result = pd.DataFrame({'Rural': Rural_value, 'Suburban': Suburban_value, 'Urban': Urban_value})

result.plot(figsize = (20,8))
plt.legend(title = "type", loc='center')

plt.xlabel("")
plt.ylabel("Fare ($USD)")
plt.title("Total Fare by City Type")

plt.savefig("Images/PyBer_fare_summary.png")
plt.show()
```
